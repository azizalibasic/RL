```python
def exp_val_of_roll(is_bad_side, value=0.0):
    exp_val = 0
    n = float(len(is_bad_side))
    p = 1.0/n
    
    for i,bad in enumerate(is_bad_side):
        m = i + 1
        if bad == 1:
            exp_val += 0.0
        else:
            n_p = p * m
            print("m={0}, n_p={1}, exp_val_p={2}".format(m,n_p,exp_val))
            exp_val += n_p
            
    return exp_val
            
```
